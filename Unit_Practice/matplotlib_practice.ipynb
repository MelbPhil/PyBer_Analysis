```python
# Import dependencies.

import matplotlib.pyplot as plt
```

```python
# Create lists for each axis.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel('Month')
plt.ylabel('Savings [$]')
plt.title('Savings Account Results')
plt.show()
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot.

plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45

plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color='blue', linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel('Date')
plt.xlabel('Fare($)')

# Create a title
plt.title('PyBer Fare by Month')

#Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

# Add titles and axes labels
ax.set_ylabel('Data')
ax.set_xlabel('Fare($)')
ax.set_title('PyBer Fare by Month')

# Invert y axis
plt.gca().invert_yaxis()

# Display legend
ax.legend()
```

```python
# Create the plot

plt.plot(y_axis, x_axis, 'o', c='red', label='Chicago')

plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')

plt.gca().invert_yaxis()

plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger =[]

for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
fix, ax = plt.subplots()
ax.scatter(x_axis, y_axis, color='blue', alpha=.2, linewidth=2, edgecolors='black', s= [i * 5 for i in y_axis], label='Boston')

ax.legend()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
explode_values = [0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
      labels=x_axis,
      autopct='%.1f%%',
      explode=explode_values,
      shadow=True,
      startangle=95,
      counterclock=False,
      colors=colors)

plt.show()
```
